```python
import os 
import shutil
import sys
birds_category_folder = "E:\\dataset\\main_project\\bird-huge\\train"
new_birds_folder = "E:\\dataset\\main_project\\bird-huge-simplified\\train"
count = 1 
os.mkdir(new_birds_folder)
```

```python
for folder in os.listdir(birds_category_folder) :
    birds_folder = os.path.join(birds_category_folder,folder) 
    for file in os.listdir(birds_folder) :  
        bird_file =  os.path.join(birds_folder,file)
        try : 
            shutil.copy(bird_file,f"{new_birds_folder}\\{count}.jpg")
            print(f"moved {bird_file} as {new_birds_folder}\\{count}.jpg")
        except Exception as e:
            print('Exception caught : ' , e) 
            sys.exit()

        count += 1 
        if count % 10 == 0 : break 

print("finished")

```
